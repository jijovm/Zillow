```python
#convert the datatype of transaction date to date and build histogram to see the 
#distribution
pd.to_datetime(merged_df.transactiondate).hist()
```

```python
#checking if there are any duplicate records on parcel ids
duplicate_df = merged_df[merged_df['parcelid'].duplicated(keep=False)]
```

```python
#number of duplicate records 
print(duplicate_df.shape)
duplicate_df.head(10)
```

The duplicate records are with different tracsaction date. We are keeping only the latest 
records

```python
#sort the records by parcel id and transaction date
unique_df = merged_df.sort_values(by=['parcelid','transactiondate'],ascending=False)
```

```python
# Remove duplicate records
unique_df.drop_duplicates(subset='parcelid',keep='first',inplace=True)
```

Checking null values

```python
unique_df.isnull().sum().sort_values(ascending=False)
```
